## Section 1.3 Analyse en Composantes Principales <a class="anchor" id="Section1.1"></a>
***

```python
# On supprime la colonne cible pour l'ACP
df_acp=df.drop(columns=('cible'))

# On renome le DF
X=df_acp
p=X.shape[1]
n=X.shape[0]
```

```python
# Instanciation
sc=StandardScaler()

# ACP sur données centrées-reduites
Z=sc.fit_transform(X)
```

```python
# Calculs des composantes principales
acp=PCA(svd_solver='full')
coord1 = acp.fit_transform(Z)
```

```python
# Nous reccuperons les données de la variable 'cible'
Quali_ue=df['cible']
modalites=np.unique(df['cible'])
```

```python
# proportion de variance expliquée
print(acp.explained_variance_ratio_)
```

****
#### Variance Expliquée
>La première composante accapare `43,18%` de l'information disponible.      
Nous disposons de `60,14%` ac les 2 premiers facteurs
>
****

```python
#liste des couleurs
couleurs = ['tomato','b']

# Création d'un graphique en coloriant les points
fig, axes = plt.subplots(figsize=(10,10))
axes.set_xlim(-4,7)
axes.set_ylim(-4,4)


# Pour chaque modalité de la var. illustrative
for c in range(len(modalites)):
#numéro des individus concernés
    numero = np.where(Quali_ue == modalites[c])
# les passer en revue pour affichage
    for i in numero[0]:
        plt.annotate(df_acp.index[i],(coord1[i,0],coord1[i,1]),color=couleurs[c],fontsize=8,alpha=0.6)
        
# ajout les axes
plt.plot([-6,7],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-6,6],color='silver',linestyle='-',linewidth=1)

plt.title('Nuage des individus dans le premier plan factoriel',fontsize=16,color='black')

plt.savefig('P10_ACP.png')

#affichage
plt.show()
```

### Le cercle des corrélations

```python
# Contribution des individus dans l'inertie totale
di=np.sum(Z**2,axis=1)
df_di = pd.DataFrame({'d_i':di})

# qualité de la représentation des individus  COS2
cos2=coord1**2
for j in range(p):
    cos2[:,j]=cos2[:,j]/di
    
# valeur corrigée
eigval = (n-1)/n*acp.explained_variance_

# Contribution aux axes
ctr=coord1**2
for j in range(p):
    ctr[:,j]=ctr[:,j]/(n*eigval[j])

df_cnt = pd.DataFrame({'CTR_1':ctr[:,0],'CTR_2':ctr[:,1]})
```

```python
# racine carré des valeurs propres
sqrt_eigval = np.sqrt(eigval)

# corrélation des variables avec les axes
corvar = np.zeros((p,p ))
for k in range(p):
    corvar[: , k] = acp.components_[k, :]*sqrt_eigval[k]

# cercle des correlations
fig,axes=plt.subplots(figsize=(8,8))
axes.set_xlim(-1,1)
axes.set_ylim(-1,1)

# Affichage des étiquettes (noms des variables)
for j in range(p):
    plt.annotate(X.columns[j],(corvar[j,0],corvar[j,1]),fontsize=10)
    plt.plot([0,corvar[j,0]],[0,corvar[j,1]],color='blue',alpha=0.2)
    
# ajouter les deux axes
plt.plot([-1,1],[0,0],color='silver',linestyle='--',linewidth=1)
plt.plot([0,0],[-1,1],color='silver',linestyle='--',linewidth=1)

# Ajouter un cercle
cercle = plt.Circle((0,0),1,color='blue',fill=False)
axes.add_artist(cercle)

plt.title('Cercle des corrélations dans le premier plan factoriel',fontsize=14,color='b')

# affichage
plt.show()
```

```python
# Création d'un df =  corrélation entre les deux premières composantes et les variables 
df_corr = pd.DataFrame({'CORR_1':corvar[:,0],'CORR_2':corvar[:,1]},index=X.columns)

# cosinus carré des variables
cos2var=corvar**2

# contributions
ctrvar = cos2var

for k in range(p):
    ctrvar[:,k] = ctrvar[:,k]/eigval[k]
    
# Création d'un DF contenant les corrélations aux axes et les leurs contributions
df_ctr = pd.DataFrame({'CTR_1':ctrvar[:,0],'CTR_2':ctrvar[:,1]},index = X.columns)
df_ctr=pd.concat([df_corr, df_ctr],axis=1,keys=["CORRELATION","CONTRIBUTION"])
df_ctr
```

****
#### Interprétation
>Nous savons, visuellement, que la prémière composante  est extrêmement  discriminante, il est donc intéressant de voir quelles sont les variables les plus coorélées et celles qui y contribuent le plus.    
Par ordre d'importance:     
**length,margin_low , margin_up , height_right**

## Partie 3 : Analyse supervisée <a class="anchor" id="Partie3"></a>
### $\;\;\;\;\;\;$Méthodes de classements

>Dans cette partie nous présentons deux méthodes de classement.   
La première méthode est la mthode des K-means et la seconde est une régression logistique.

## Section 3.1 Données d'apprentissage et données de test <a class="anchor" id="Section3.1"></a>
###  $\;\;\;\;\;\;$Création de  'set_training &  set_test'

```python
df['cible'] = np.where(df['cible']=='true',1,0)

X = df_acp
y = df['cible']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## Section 3.2 Méthode des K-means  <a class="anchor" id="Section3.2"></a>

**Présentation de la méthode:**      
>Il s'agit de créer deux groupes, la variable 'cible' étant la référence.      
Ensuite, pour chaque nouvelle donnée, le modèle calcule la distance au centre le plus proche.  

```python
# Nous travaillons avec les données centrées-reduites
data_c = preprocessing.StandardScaler().fit(X_train)
X_cr = data_c.transform(X_train)

# K-means sur les donnée d'apprentissage centrées et reduites 
kmeans_cr =KMeans(n_clusters=2, random_state=0).fit(X_cr)
```

```python
# distances aux centres de classes pour chaque individu
df_k=pd.DataFrame(kmeans_cr.transform(X_cr))

# On met les plus petites valeurs par ligne (distance au centre des goupes) en couleur bleu
def highlight_min(s, props=''):
    return np.where(s == np.nanmin(s.values), props, '')

df_k = df_k.style.apply(highlight_min, props='color:white;background-color:darkblue', axis=1)
```

```python
# Impoortation de la matrice de confusion

from sklearn.metrics import confusion_matrix
confusion_1 = pd.DataFrame(confusion_matrix(y_train, kmeans_cr.labels_))
confusion_1
```

```python
# Nous travaillons avec les données centrées-reduites
data_c = preprocessing.StandardScaler().fit(X_test)
X_cr = data_c.transform(X_test)

# Création d'une liste contenant les prédictions
test = kmeans_cr.fit_predict(X_test) 

confusion_2 = pd.DataFrame(confusion_matrix(y_test, test))
confusion_2
```

****
>Au vu des résultats de l'ACP il est normal que les données soient bien scindable en deux groupes, nous rappelons que la méthode des K-means est une méthode géométrique.      
Notre modèle prédictif des K-means donnent de très bon résultat.       
Avec les résultats de la matrice de confusion des données de test nous concluons qu'il n'y a pas de sur-apprentissage du modèle.
****

## Section 3.3 Régression logistique  <a class="anchor" id="Section3.3"></a>

### $\;\;\;\;\;\;$3.3.1) Modèle complet avec constante <a class="anchor" id="Section3.3.1"></a>

```python
# Nous ajoutons une constante à X_train et X_test
from statsmodels.tools import add_constant
X_train= sm.tools.add_constant(X_train)
X_test= sm.tools.add_constant(X_test)
```

```python
# Importation de la classe de calcul
from statsmodels.api import Logit

# Regression logistique 
lr=Logit(endog=y_train,exog=X_train)

# Lancement des calculs
# Algorithme de Newton-Raphson utilisé par defaut
res=lr.fit()
```

```python
# résumer des resultats
print(res.summary())
```

```python
# Matrice de confusion
confusion_3 = pd.DataFrame(res.pred_table())
confusion_3
```

### Indicateur de performance de l'échantillon test
####  $\;\;\;\;\;\;$ Modèle complet

#### a) Matrice de confusion   <a class="anchor" id="Section3.3.1.a"></a>

```python
# Test-Model
prediction = res.predict(X_test)
predSm = np.where(prediction>=0.5,1,0)

# matrice de confusion des données de test
confusion_4 = pd.crosstab(y_test,predSm)
confusion_4
```

#### b)Taux de succès et taux d'erreur <a class="anchor" id="Section3.3.1.b"></a>

```python
# Taux de reconnaissance
mc = confusion_4.values
accSm = np.sum(np.diagonal(mc))/np.sum(mc)
print("Taux de succès : ",round(accSm,3))

# Taux d'erreur
errSm = 1 - accSm
print("Taux d'erreur : ",round(errSm,3))
```

#### c) Courbe ROC <a class="anchor" id="Section3.3.1.c"></a>

```python
# Importer le module metrics de la librairie scikit-learn
import sklearn.metrics as metrics

fprSm, tprSm, _ = metrics.roc_curve(y_test , predSm, pos_label = 1)

# Construire la diagonale de référence
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1),'b')

# Rajouter notre diagramme
plt.plot(fprSm,tprSm,'r')

plt.title("Courbe ROC")
plt.savefig('P10_ROC_1.png')
plt.show()
```

#### d) Calcul de l'AUC <a class="anchor" id="Section3.3.1.d"></a>

>Pour calculer l'AUC (l'aire sous la courbe) nous utilisons la fonction **roc_auc_score()**

```python
# Calcul de l'AUC
aucSm = metrics.roc_auc_score(y_test,predSm)
print("AUC :",round(aucSm,3))
```

#### Conclusion à propos du modèle de régression logistique complet
>Nous avons un modèle de classement presque parfait le taux d'erreur est de 7%, c'est à dire que 93% des nouvelles données sont bien classées.    
La courbe ROC est à l'image d'un modèle parfait et l'AUC est proche de 1.

>Voyons si nous pouvons améliorer le caractère prédictif du modèle en retirant les deux variables non-significatives:    
La constante et la longueure de la diagonale.

### $\;\;\;\;\;\;$3.3.2) Modèle épuré <a class="anchor" id="Section3.3.2"></a>

>Dans cette partie de l'étude, nous allons relancer le modèle, mais cette fois-ci, sans les variables non-significative.

```python
# Training-Model

# Suppression des deux variables non-significative
X_train =  X_train.drop(columns=['const','diagonal'])
X_test  =  X_test.drop(columns=['const','diagonal'])

# Regression logistique
lr=Logit(endog=y_train,exog=X_train)

# Lancement des calculs
# Algorithme de Newton-Raphson utilisé par defaut
res=lr.fit()
```

```python
# Affichage des résultats
print(res.summary())
```

```python
# Matrice de confusion
confusion_5 = pd.DataFrame(res.pred_table())
confusion_5
```

### Indicateur de performance de l'échantillon test
####  $\;\;\;\;\;\;$ Modèle épuré

#### a) Matrice de confusion   <a class="anchor" id="Section3.3.2.a"></a>

```python
# Test-Model
prediction = res.predict(X_test)
predSm = np.where(prediction>=0.5,1,0)

# matrice de confusion des données de test
confusion_6 = pd.crosstab(y_test,predSm)
confusion_6
```

#### b) Taux de succès et taux d'erreur <a class="anchor" id="Section3.3.2.b"></a>

```python
# Taux de reconnaissance
mc = confusion_4.values
accSm = np.sum(np.diagonal(mc))/np.sum(mc)
print("Taux de succès : ",round(accSm,3))

# Taux d'erreur
errSm = 1 - accSm
print("Taux d'erreur : ",round(errSm,3))
```

#### c) Courbe ROC <a class="anchor" id="Section3.3.2.c"></a>

```python
# Importer le module metrics de la librairie scikit-learn
import sklearn.metrics as metrics

fprSm, tprSm, _ = metrics.roc_curve(y_test , predSm, pos_label = 1)

# Construire la diagonale de référence
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1),'b')

# Rajouter notre diagramme
plt.plot(fprSm,tprSm,'r')

plt.title("Courbe ROC")
plt.savefig('P10_ROC_2.png')
plt.show()
```

#### d) Calcul de l'AUC <a class="anchor" id="Section3.3.2.d"></a>

>Pour calculer l'AUC (l'aire sous la courbe) nous utilisons la fonction **roc_auc_score()**

```python
# Calcul de l'AUC
aucSm = metrics.roc_auc_score(y_test,predSm)
print("AUC :" ,round(aucSm,3))
```

#### Conclusion à propos du modèle de régression logistique épuré
>Nous reccupérons les mêmes résultats:    
le taux d'erreur est de 7%, c'est à dire que 93% des nouvelles données sont bien classées.    
La courbe ROC est à l'image d'un modèle parfait et l'AUC est proche de 1.

>Il n'y a pas d'amélioration en retirant les deux variables

##  Conclusion & Recommandations<a class="anchor" id="Conclusion"></a>

>Nous venons de voir deux méthodes de classement les K-means et la regression logistique.     
Les deux méthode donne exactement les mêmes résultats. Aucune des deux méthodes ne présentent de difficulté pour sa mise en place.    

>Notons que la méthode des K-means ne possède pas de "Selection de variables" automatisé, par opposition à la régression logistique,  en cas de modèle complexe, discrimination pénible à réaliser, les modules qui composent les classes de la regression logistique feront la différence.

>Quel que soit l'étude il est toujours utile de confronter plusieurs modèle et de selectionner le meilleur. Ici le praticien peut utiliser le modèle de son choix.
